### Groupby method

```python
df.head()
```

```python
df.head(2)
```

```python
# drop col 'LAT_DEG', column name
df.drop(['LAT_DEG','LONG_DEG'], axis=1)

# change df in-place
df.drop(['LAT_DEG','LONG_DEG'], axis=1, inplace = True)  
```

```python
df.head()
```

```python
# group by value, calculate mean
df.groupby('UNION').mean()
```

```python
df.groupby('UNION').mean().plot()
```

```python
%pip install ipympl nodejs-bin
```

```python
# make it interactive
#import matplotlib_inline
%matplotlib inline
df.groupby('UNION').mean().plot(kind='bar')
```

```python
# groupby two columns for all values
df.groupby(['UNION','WELL_DEPTH']).mean()
```

```python
# groupby and get values of single col
df.groupby(['UNION'])['Ca'].mean()
```

```python
# groupby column, plot bar chart
df.groupby('UNION')['Ca'].agg(['mean', 'std']).plot(kind='bar')
```

```python
# rename column
df.rename(columns = {'UNION':'Location'}, inplace=True)
```

```python
df
```

```python
# groupby for count
df.groupby('Location')['Ca'].count()
```

```python
df.groupby('Location')['Ca'].count().plot()
```

```python
# plot using matplotlib, rather inline
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
df.groupby('Location')['Ca'].count().plot(kind='bar')
```

```python
# add new column to df
df['Ca_percentage'] = df.Ca/sum(df.Ca)*100
```
