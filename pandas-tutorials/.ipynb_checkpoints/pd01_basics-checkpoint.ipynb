## Indexing and slicing

```python
df['UNION']  # extract single col
```

```python
df[['UNION','WELL_DEPTH']]  # use two [[]], show 2 cols, 
```

```python
# df dot the column - another way of indexing
df.UNION, df.Ca
```

```python
# indexing using iloc, or loc, locator acessor
# index, slice, 
df.iloc[2,2]
```

```python
# Union index
# 2:5 - from 2nd row to 6th row, row size
# 0:4 - from 0 col to 4th col, column size
df.iloc[2:6, 0:5]
```

```python
# in python, 0 can be ignored, [,5] = [0,5]
df.iloc[2:6, :5]
```

```python
# from 3rd index till end
df.iloc[2:6,3:]
```

```python
# [:] - all rows
df.iloc[:,3:]
```

```python
# Use colomn name index
df.loc[2:10,['HCO3', 'Mg']]
```

```python
# Add filter condition for a column
df.loc[df['Ca']>88]
```

```python
# return multiple columns
df.loc[df['Ca']>88, ['Ca', 'UNION']]
```

```python
# Multiple conditions
df.loc[(df['Ca']>88) & (df['Ca']>120)]
```

```python
# Retrieve values from data
df.loc[df.UNION=='Aliabad']
```

```python
# Retrieve string values
df.UNION.str.contains('Aliabad')
```

```python
# regex values start with A
df[df.UNION.str.contains('^A')]
```
