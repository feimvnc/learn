### Tidy messy datasets  
Common problems  
Column headers are values, not variable names  
Multiple variable are stored in one column  
Variables are stored in both rows and columns  
Multiple types of observational units are stored in the same table  
A single observational unit is stored in multiple tables  

https://github.com/hadley/tidy-data


```python
import pandas as pd
df = pd.read_csv('pew.csv')
df
# below header are vavlues, change header values to a column
```

```python
df.melt(id_vars='religion', var_name='income_range', value_name='count')
```

### Another dataset  


```python
ebola = pd.read_csv('country_timeseries.csv')
ebola
# country names are in header
```

```python
#
df = ebola.melt(id_vars = ['Date', 'Day'], var_name='Country', value_name='count')
df
```

```python
#plot
df.groupby('Country')['count'].mean().plot(kind = 'bar', figsize = (12, 6))
```
