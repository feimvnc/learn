```python
# Height to numeric, and clean up
def height_to_numerical(height):
    """
    Convert string to total inches
    Ex. '5-11' --> 71
    """
    feet = height.split('-')[0]
    inches = height.split('-')[1]
    return int(feet)*12 + int(inches)

def clean_height(val):
    try:
        # height is always in inches
        height = int(val)
    except:
        height = height_to_numerical(val)

    return height
```
