### Missing values

```python
import pandas as pd
import numpy as np
%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train price.csv')
df
```

```python
# sample small set with missing values
df = pd.DataFrame([[1,2,3,4],[5,6,7,np.nan],[8,9,np.nan,np.nan],], \
                  columns=['A','B','C','D'])
df
```

```python
# first find nan values
df.isna()
```

```python
# count nan
df.isna().sum()
```

```python
# drop nan colum
df.dropna(axis = 'columns')
# df.dropna(axis = 'columns', inplace=True)  # modify df
```

```python
# fill nan value
df.fillna(df.mean())
```

```python
# fill with previous value
df.ffill()
```

```python
# fill with next value
df.bfill()
```

```python
# replace 5 with 10
df.replace(5, 10)
```

```python
# load sample data
df = pd.read_csv('train price.csv')
df.shape
```

```python
msk = (df.isna().sum()/len(df)*100).sort_values(ascending=False)[:20]
msk
```

```python
# filter msk > 10
msk = msk[msk>10]
msk
```

```python
# show index names
msk = msk[msk>10].index
msk
```

```python
# drop index with msk > 10
df = df.drop(df[msk], axis=1)
df
```

```python
# col numbers changed from 81 to 74
df.shape
```

```python
# calcualte threshold
1460 * .9
```

```python
# Use threshold values of nan counts, same effects as above
d = df.dropna(axis = 1, thresh = 1314)
d.shape
```

### Impute or replace values

```python
for col in ('GarageType','GarageFinish','GarageQual','GarageCond'):
    df[col] = df[col].ffill()
df.head(3)    
```

### Duplicates  
dupliates are rows that occur repeatedly in dataset

```python
df.duplicated().sum()  # no duplicate
```

```python
# manually create duplicated rows, rows 0-15, all columns
df2 = df.loc[0:15, :]
newdf = pd.concat((df, df2))
newdf
```

```python
# check duplicates, total 15
newdf.loc[newdf.duplicated()]
```

```python
newdf.shape
```

```python
#drop duplicates
newdf.drop_duplicates(inplace = True)  # show 1476 - 16 = 1,460 rows
```
