```python
import pandas as pd
df = pd.read_parquet('stations_with_stats.parquet')
df
```

```python
df_rain = df[['Series Start', 'Yearly rainfall']]
df_rain = df_rain.rename(columns={"Series Start": "Year", "Yearly rainfall": "Rainfall"})
df_rain
```

```python
df_elev = df[['Series Start', 'Elevation']]
df_elev = df_elev.rename(columns={"Series Start": "Year"})
df_elev
```

```python
# Join two df
# 'outer' join adds nan if no value available, like union
df_merged = pd.merge(df_rain, df_elev, on = 'Year', how = 'outer')
df_merged
```

```python
# inner join not having nan, only include common values
newdf =pd.merge(df_rain, df_elev, on = 'Year', how = 'inner')
newdf.isnull().sum()
```
