```python
import pandas as pd
df = pd.read_excel('Water quality.xlsx')
df.head()
```

```python
# rename column name
df.rename(columns = {'UNION':'Location'}, inplace=True)
```

```python
# add a new col with len of Location index
df['len'] = df.Location.apply(len)
df
```

```python
# check index value len
df.Location.str.len()
```

```python
# use function or lambda do do apply

def check_concentration(con):
    if con < 10: return 'good'
    elif (con > 10) & (con <= 50): return 'pool'
    else: return 'unsuitable'   # unsuitable to drink

df.AsIII.apply(check_concentration)
```

```python
# add a new column
df['water_q'] = df.AsIII.apply(check_concentration)
df
```

```python
# lambda method - 1 to apply
df.AsIII.apply(lambda con: 'good' if con<10 else 'pool')
```

```python
# lambda method - 2
df.AsIII.apply(lambda con: 'good' if con<10 else ('pool' if con>10 & con<=50 else 'unsuitable'))
```

```python
df['lambda'] = df.AsIII.apply(lambda con: 'good' if con<10 else ('poor' if (con>10 & con<=50) else 'unsuitable'))
df
```

```python
# drop string index, get mean
df.drop(['Location','water_q', 'lambda'], axis=1).apply(lambda x: x - x.mean())
```

```python
#replace value 990 with Nan
import numpy as np
df.replace([990, 7], np.nan, inplace=True)
df
```

```python
# now fill Nan with mean value
df.groupby('Location')['AsIII'].apply(lambda x: x.fillna(x.mean()))
```
